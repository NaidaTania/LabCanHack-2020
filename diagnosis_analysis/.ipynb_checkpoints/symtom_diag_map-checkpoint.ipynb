```python
import pandas as pd 
import numpy as np 
```

```python
symptoms = pd.read_csv('sym_t.csv')
symptoms = symptoms.set_index('syd')
print(symptoms.head())
print(symptoms.shape)
print(symptoms.at[1,'symptom'])
```

```python
diagnosis = pd.read_csv('dia_t.csv')
diagnosis = diagnosis.set_index('did')
print(diagnosis.shape)
diagnosis.head()
```

```python
match = pd.read_csv('diffsydiw.csv')
match = match.set_index('did')
match = match.drop('wei', axis = 1)
print(match.shape)
match.head(20)
```

```python
print(match.head())
column = []
count = 0
for i, row in match.iterrows(): 
    if not np.isnan(row['syd']):
        count += 1
        column.append(symptoms.at[int(row['syd']),'symptom'])
    else: 
        column.append('')
match['syd_t'] = column
print("count:", count)
match.head()
```

```python
mat = {}

for i, row in match.iterrows(): 
    if row['syd_t'] == 'NaN': 
        continue
    if i in mat: 
        mat[i].append(row['syd_t'])
    else: 
        mat[i] = [row['syd_t']]
        
print(type(mat))
diag_symp = {}
for key in mat:
    if not np.isnan(key):
        diag_symp[diagnosis.at[int(key),'diagnose']] = mat[key]

diag_symp["Constipation"]
diag_symp.keys()
```
